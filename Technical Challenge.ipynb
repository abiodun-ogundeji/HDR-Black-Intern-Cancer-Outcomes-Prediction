```python
import pandas as pd
```

```python
patient_train = pd.read_csv('av_patient_train.csv')

patient_train = patient_train.drop(patient_train.columns[0], axis=1)
print(patient_train.shape)
patient_train.head()
```

```python
patient_train.columns
```

```python
tumour_train = pd.read_csv('av_tumour_train.csv', low_memory=False)

tumour_train = tumour_train.drop(tumour_train.columns[0], axis=1)
print(tumour_train.shape)
tumour_train.head()
```

```python
tumour_train.columns
```

```python
# Merge the dataframes based on multiple columns (ID, Name)
train_data = pd.merge(patient_train, tumour_train, on=['PATIENTID', 'SEX', 'LINKNUMBER'])

print(train_data.shape)
train_data.head()
```

```python
patient_test = pd.read_csv('av_patient_test.csv')

patient_test = patient_test.drop(patient_test.columns[0], axis=1)
print(patient_test.shape)
patient_test.head()
```

```python
patient_test.columns
```

```python
tumour_test = pd.read_csv('av_tumour_test.csv', low_memory=False)

tumour_test = tumour_test.drop(tumour_test.columns[0], axis=1)
print(tumour_test.shape)
tumour_test.head()
```

```python
# Merge the dataframes based on the 'ID' column
test_data = pd.merge(patient_test, tumour_test, on=['PATIENTID', 'SEX', 'LINKNUMBER'])

print(test_data.shape)
test_data.head()
```
